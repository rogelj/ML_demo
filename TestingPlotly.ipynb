# Using Plotly

A nice open source library fro interactive viz

```python
from plotly import __version__
from plotly.offline import download_plotlyjs, init_notebook_mode, iplot

print(__version__) # requires version >= 1.9.0
```

```python
init_notebook_mode(connected=True) # run at the start of every ipython notebook to use plotly.offline
                     # this injects the plotly.js source files into the notebook
```

```python
from plotly.graph_objs import Scatter, Figure, Layout
```

```python
iplot([{"x": [1, 2, 3], "y": [3, 1, 6]}])
```

```python
from plotly.graph_objs import *
import numpy as np
```

```python
# Scatter plot with heatmap
x = np.random.randn(2000)
y = np.random.randn(2000)
iplot([Histogram2dContour(x=x, 
                          y=y, 
                          contours=Contours(coloring='heatmap')),
       Scatter(x=x, y=y, mode='markers', 
               marker=Marker(color='white', 
                             size=3, 
                             opacity=0.3))],
      show_link=False)
```
