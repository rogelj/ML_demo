# Scikit

Using the Iris data set.

```python
def get_num(x):
    if x=='setosa':
        y=0
    elif x=='versicolor':
        y=1
    else:
        y=2
    return y
```

```python
iris_data['target']= iris_data['species'].apply(get_num)
```

```python
iris_data.tail()
```

## Modelling the data


```python
X = iris_data[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']].values
Y = iris_data['target'].values
```
